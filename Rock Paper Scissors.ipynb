```python
def rock_paper_scissors():
    import random
    choices = ['Rock', 'Paper', 'Scissors']
    c_choice = ""
    p_score = 0
    c_score = 0
    p_name = input("Player name: ")
    
    while True:
        c_choice = random.choice(choices)
        print(p_name, "has", p_score, "points.")
        print("Computer has", c_score, "points")
        print("Type \"Exit\" to quit.")
        print("")
        p_choice = input("Rock, Paper or Scissors? ")
        print("")

        
        try:
            if p_choice == "Exit":
                if p_score > c_score:
                    print("EZ DUB")
                    break
                elif p_score < c_score:
                    print("You suck")
                    break
                elif p_score == c_score:
                    print("Tie game.")
                    break
                
            elif p_choice == c_choice:
                print("Computer picked", c_choice)
                print("Tie game.")
                print("")
                
            elif p_choice == "Rock":
                if c_choice == "Paper":
                    print("Computer picked", c_choice)
                    print("Computer wins!")
                    print("")
                    c_score += 1
                elif c_choice == "Scissors":
                    print("Computer picked", c_choice)
                    print(p_name, "wins!")
                    print("")
                    p_score += 1
                    
            elif p_choice == "Paper":
                if c_choice == "Scissors":
                    print("Computer picked", c_choice)
                    print("Computer wins!")
                    print("")
                    c_score += 1
                elif c_choice == "Rock":
                    print("Computer picked", c_choice)
                    print(p_name, "wins!")
                    print("")
                    p_score += 1
                    
            elif p_choice == "Scissors":
                if c_choice == "Rock":
                    print("Computer picked", c_choice)
                    print("Computer wins!")
                    print("")
                    c_score += 1
                elif c_choice == "Paper":
                    print("Computer picked", c_choice)
                    print(p_name, "wins!")
                    print("")
                    p_score += 1
                      
        except:
            print("Input error")
            print("")
```
